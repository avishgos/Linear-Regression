# Feature Selection

# Recursive Feature Elimination

```python
df = df.drop('car_name',axis = 1)
```

```python
y = df['mpg']
X = df.drop('mpg',axis = 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import r2_score,mean_squared_error
```

```python
lr = LinearRegression()
rfe = RFE(lr,n_features_to_select=9)
rfe.fit(X,y)

#WE ASKED ALGORITHM TO TELL US WHICH FEATURE IS WORSE BY PUTTING 'n_features_to_select = 9
```

```python
rfe.support_
```

```python
pd.DataFrame(rfe.ranking_,index = X.columns,columns = ['Select'])
```

This will give 7 best features with rank 1 and rest features with subsequent ranks.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=3)
```

```python
no_of_cols = X_train.shape[1]
r2score = []
rmse = []

lr = LinearRegression()   #estimator

for i in range(no_of_cols):
    rfe = RFE(lr,n_features_to_select=i+1)
    #fit rfe on the train set
    rfe.fit(X_train,y_train)
    y_test_pred = rfe.predict(X_test)
    
    #for r2score 
    r2 = r2_score(y_test,y_test_pred)
    r2score.append(r2)
    
    #now for mean square error
    rms = np.sqrt(mean_squared_error(y_test,y_test_pred))
    rmse.append(rms)
    
```

```python
plt.plot(range(1,8),r2score)
```

```python
r2score
```

```python
plt.plot(range(1,8),rmse)
```

```python
rmse
```

# Hyperparameter Tuning Approach to find the optimal number of features to keep

```python
from sklearn.model_selection import KFold,GridSearchCV

params = {'n_features_to_select': list(range(1,8))}
lr = LinearRegression()
rfe = RFE(lr)

kf = KFold(n_splits=3,random_state=3)

gsearch = GridSearchCV(rfe,param_grid = params, cv = kf,return_train_score=True,scoring='r2')
gsearch.fit(X,y)
```

```python
gsearch.best_params_   #we found that if we keep n = 7,its helping us
```

```python
pd.DataFrame(gsearch.cv_results_)
```

# Forward selection to decide the best number of features to keep in the model(STEP FORWARD SELECTION - SFS)

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
```

```python
lr = LinearRegression()
sfs1 = sfs(lr,k_features=7,scoring='r2',cv = 3,verbose=2)
sfs1.fit(X,y)
```

```python
sf = pd.DataFrame(sfs1.subsets_).T
sf
```

```python
plt.plot(sf['avg_score'])
plt.show()
```

```python
cols = list(sfs1.k_feature_names_)
cols
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[cols], y, test_size=0.3, random_state=3)

lr.fit(X_train, y_train)
y_test_pred = lr.predict(X_test)

plt.scatter(y_test, y_test_pred)
plt.plot(y_test, y_test,'r')
```
