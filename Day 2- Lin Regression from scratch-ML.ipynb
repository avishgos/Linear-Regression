```python
#importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

```python
df = pd.read_csv('airline cost new.csv')
df.head()
```

```python
y = df['cost']
X = df['num_pass']
```

```python
plt.scatter(X,y)
plt.show()
```

```python
#Formulation of Gradient Descent
m = 30
c = 1200

L = 0.0001
epochs = 1000

n = float(len(X))
for i in range(epochs):
    y_pred = m*X+c
    D_m = (-2/n)*sum(X*(y-y_pred))
    D_c = (-2/n)*sum(y-y_pred)
    
    m = m-D_m*L
    c = c-D_c*L
    
    if (i%100)==0:
        plt.scatter(X,y)
        plt.plot(X,y_pred, 'r')
        plt.show()
        
print(m ,c)
```
